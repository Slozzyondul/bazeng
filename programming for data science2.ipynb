```python
#loading an excel file
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x) #rounds the number to 2 dp
df = pd.read_csv('/home/solo/Desktop/mls_salaries1.csv', index_col=None)
df.head(4)
df['club']
df[['club', 'base_salary', 'position']]
#filtering
millionaires = df['base_salary'] > 1000000
df[millionaires]
```

```python
#EDA 
df.tail()
df.head()
df.info()
df.describe()
df.describe() #df.describe().round()
df.describe().apply(lambda s: s.apply('{0:.0f}'.format)) #applies this format to the entire return   
df['difference'] = df['guaranteed_compensation'] - df['base_salary'] #creating new column with an operation
df.head()

#clean and manipulte data
df.isnull().sum()

#drop missing values
df.dropna(inplace=True)
df.isnull().sum()

def clean_currency(x):
    """if the value is a string, remove currency symbol and delimiters. Otherwise,the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)

df['base_salary'] = df['base_salary'].apply(clean_currency).astype('int')
#print(df['base_salary'])
df['base_salary']

#Data visualization with pandas
import pandas as pd
import matplotlib.pyplot as plt
df.head()
```
