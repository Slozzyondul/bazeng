```python
#Exploratory data analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/home/solo/Desktop/Programming for Data Science/python-pandas-for-data-manipulation/completed_files/datasets/mtcars.csv')
df

#sns.pairplot(df)
```
