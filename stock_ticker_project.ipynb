```python
#creating a table with mutiple column representing various tag values
product_name = soup.find_all('a', class_= 'title')
product_name

#product prices
product_price = soup.find_all('h4', class_ = 'float-end price card-title pull-right')
product_price

#get product reviews
import re
product_review =soup.find_all('p', class_ = re.compile('float-end review-count'))
product_review

#get product description
product_description = soup.find_all('p', class_ = re.compile('description card-text'))
product_description

#create a for loop to gather everything
#name of the item
product_name_list = []
for i in product_name:
    name = i.text
    product_name_list.append(name)
product_name_list    

#price of the item
products_price_list = []
for i in product_price: 
    price = i.text
    products_price_list.append(price)
products_price_list

#item reviews
product_review_list = []
for i in product_review:
    review = i.text
    product_review_list.append(review)
product_review_list

#description list
product_description_list = []
for i in product_description:
    description = i.text
    product_description_list.append(description)
#product_description_list

#converting it into a dataframme
import pandas as pd
table = pd.DataFrame({'Product Name':product_name_list, 'Product Price':products_price_list,'Product Review':product_review_list, 'Product Description':product_description_list})

                      
table
table.info()
table.describe()
```
