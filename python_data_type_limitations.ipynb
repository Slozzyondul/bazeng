```python
#Social Network Analysis
import networkx as nx

G = nx.Graph()
G.add_nodes_from(["Alice", "Bob", "Carlie", "David", "Eve"]) 
G.add_edge("Alice", "Bob")
G.add_edge("Bob", "Carlie")
G.add_edge("Carlie", "David")
G.add_edge("David", "Eve")
G.add_edge("Eve","Alice")
print(G.edges())

import matplotlib.pyplot as plt
nx.draw_networkx(G, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=nx.spring_layout(G), labels=labels)
pos = nx.spring_layout(G)

node_labels = nx.get_node_attributes(G, 'label')
for node, label in node_labels.items():
    x,y = pos[node]
    plt.text(x, y, label, fontsize=12)

plt.show()
```

```python
#Complexity analysis and memory O(n) fair
list = [1,2,3,4,5,2]
#finding the maximum number in a list
def find_max(list):
    max_value = list[0]
    for value in list:
        if value > max_value:
            max_value = value
    return max_value
find_max(list)        

#finding duplicates in a list o(n*n)

def check_duplicates(list):
    for i in range(len(list)):
        for j in range(i+1, len(list)):
            if list[i] == list[j]:
                return True
    return False
check_duplicates(list)        
```

```python
# 0(1) constant time: time is not affected by input size
def find_element(list, index):
    return list[index]

#0(log n) Logarithmic time: run time increases logarithimically with the input size
def binary_search(list, target):
    left = 0
    right = len(list) - 1
    while left <= right:
        mid = (left+right) // 2
    
        if list[mid] == target:
            return mid
        elif list[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1            


# O(n) linear time: running of the algorithim increases lenearly based on the input
def find_max(list):
    max_val = list[0]
    for i in range(1, len(list)):
        if list[i] > max_val:
            max_val = list[i]
    return max_val      


# O(n**2) Quadratic time: run time increases at the rate of n*n
def find_pairs(list, target):
    pairs = []
    for i in range(len(list)):
        for j in range(1+1, len(list)):
            if list[i] + list(j) == target:
                pairs.append((list[i], list[j]))
    return pairs         

#  O(2^n) exponential time: run time increases exponentially based on the input
def subsets(list):
    if len(list) == 0:
        return[[]]
    x = subsets(list[1:])
    return x +[[list[0]] +y for y in x]


# O(n!) factorial time
def permutation(string):
    if len(string) == 0:
        return['']
    prev_list = permutation(string[1:])
    next_list = []
    for i in range(len(prev_list)):
        for j in range(len(prev_list[i])+1):
            new_string = prev_list[i][:j]+string[0]+prev_list[i][j:]
            if new_string not in next_list:
                next_list.append(new_string)
    return next_list          
permutation("Solomon")  
```
